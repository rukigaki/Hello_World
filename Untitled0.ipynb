```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/hh.csv', index_col=0)
```

```python
for i in range(df.values.shape[1]):
  for j in range(df.values.shape[0]):
    df.values[j][i] = df.values[j][i].replace("\n", "")
    df.values[j][i] = df.values[j][i].replace("\xa0", " ")
```

```python
arr = []
for i in df.values[:, 4]:
  res = i.split(',')
  for j in res:
    if j not in arr:
        arr.append(j)
arr        
```

```python
def getParameterCity(arg):
  millionCity = "Новосибирск Екатеринбург Нижний Новгород Казань Челябинск Омск Самара Ростов-на-Дону Уфа Красноярск Пермь Воронеж Волгоград Сочи"
  sarg = arg.split(',')
  print(sarg)
  item = sarg[0]
  item = item.strip()
  if item == "Москва":
    return "Москва"
  elif item == "Санкт-Петербург":
    return "Санкт-Петербург"
  elif item in millionCity:
    return item
  else:
    return "Малый или за пределами России город"


getParameterCity(df.values[0][3])
```

```python
def getParameterEmployment(arg):
  result = ""
  if (arr[3] in arg) | (arr[-1] in arg) | (arr[-3] in arg):
    result += arr[3]
  if (arr[1] in arg) | (arr[-2] in arg):
    result += arr[1]
  if (arr[2] in arg) | (arr[4] in arg):
    result += arr[2]
  if (arr[0] in arg) | (arr[5] in arg):
    result += arr[0] 
  if result == "":
    result += "Не указано" 
  return result  

getParameterEmployment(df.values[7, 4])    
```

```python
arr = []
for i in df.values[:, 5]:
  s = i.split(',')
  for j in s:
    if j not in arr:
      arr.append(j)

arr  
```

```python
def getParameterSchedule(arg):
  result = ""
  if (arr[1] in arg) | (arr[5] in arg):
    result += arr[1]
  if (arr[0] in arg) | (arr[6] in arg):
    result += arr[0]
  if (arr[4] in arg) | (arr[-3] in arg):
    result += arr[4]
  if (arr[2] in arg) | (arr[-1] in arg):
    result += arr[2]
  if arr[-2] in arg:
    result += " ротационная работа"  
  if result == "":
    result += "Не указан"  

  return re.sub(r'\s',result  


getParameterSchedule(df.values[1, 5])
```
