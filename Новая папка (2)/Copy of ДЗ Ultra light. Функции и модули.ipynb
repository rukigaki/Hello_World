**Задача 1**

Напишите функцию, которая будет выводить количество букв в слове. Проверьте при помощи этой функции, сколько букв в слове восьмидесятивосьмимиллиметровое.

**Решение:**

```
def letters_in_a_word(word):
  return len(word)

letters_in_a_word("восьмидесятивосьмимиллиметровое")
```

**Задача 2**

Напишите функцию, которая выводит таблицу умножения (от 1 до 9) для указанного числа в следующем формате:

1 * 3 = 3

2 * 3 = 6

3 * 3 = 9 

...

9 * 3 = 27

**Решение:**

```
def multiplication_tables_for_three(x):
  for i in range(1, 10):
    print(str(i), "*", str(x), "=", str(i*x))
multiplication_tables_for_three(3)    
```

**Задача 3**

Напишите функцию, которая выводит сумму, разность и произведение двух, поданных на вход чисел.


**Решение:**

```
def a(a, b):
  x = ["Сумма:", "Разность:", "Произведение:"]
  def a2(y = []):
      y.append(a+b)
      y.append(a-b)
      y.append(a*b)
      d = dict(zip(x, y))
      return d
  for i in a2():
    print(i, a2()[i]) 
a(1, 2)    
```

**Задача 5**

Напишите лямбда-функцию, которая считает куб числа.



**Решение:**

```
def cube(x):
  return (lambda y: y**3)(x)
cube(2) 
```
